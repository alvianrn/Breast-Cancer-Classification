# Undersampling Data

Undersampling adalah salah satu teknik dalam penyeimbangan data (data balancing) yang digunakan untuk menangani masalah ketidakseimbangan kelas (class imbalance) dalam dataset. Ketidakseimbangan kelas terjadi ketika satu atau lebih kelas dalam data klasifikasi memiliki jumlah contoh yang jauh lebih sedikit atau lebih banyak dibandingkan kelas lainnya.

```python
dataset = '/content/drive/MyDrive/Dataset/Balance Breast Cancer Dataset'
os.listdir(dataset)
```

```python
list = []
totalFiles = 0
totalDir = 0
for base, dirs, files in os.walk(dataset):
    list.append(len(files))
    #print('Searching in : {} with total {} image'.format(base,len(files)))
    for directories in dirs:
        totalDir += 1
    for Files in files:
        totalFiles += 1
```

```python
d = {'Class': os.listdir(dataset), 'Total_image_in_class': np.delete(list, [0])}
data1 = pd.DataFrame(data=d)
data1
```

```python
fig, ax = plt.subplots(figsize=(12,6))
sns.barplot(data=data1,x='Class', y='Total_image_in_class')
plt.title('Class distribution')
ax.bar_label(ax.containers[0])
ax.grid()
plt.show()
```

```python
normal_dir = os.path.join(dataset, 'normal')
malignant_dir = os.path.join(dataset, 'malignant')
benign_dir = os.path.join(dataset, 'benign')
```

```python
import random
import os

path = malignant_dir
n = 30
img_names = os.listdir(path)
img_names = random.sample(img_names, n)
for image in img_names:
  f = os.path.join(path, image)
  os.remove(f)
```

```python
path = benign_dir
n = 257
img_names = os.listdir(path)
img_names = random.sample(img_names, n)
for image in img_names:
  f = os.path.join(path, image)
  os.remove(f)
```
